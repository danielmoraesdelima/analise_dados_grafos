#### Carteiro Chines - Caminho ou Circuito de Euler

```python
# Outro Exemplo de utilização de Grafos Eulerianos
# Criação do Grafo
G = nx.Graph()

# Criação de toas as arestas, com apoio de outros colegas que já haviam desenhando os vértices.
E = [(1,2), (1,3), (1,6), (1,8), (2,4), (2,5), (2,7), (3,8), (4,7), (5,7), (6,8), (7,8)]

# Adicionando as Arestas ao Grafo
G.add_edges_from(E)

# Mostra o grafo colorido
G.add_edges_from(E)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color='b')

# Verificar se o Grafo é Eureliano
print('Grafo G é Eureliano?',nx.is_eulerian(G))

# Mostra o grafo colorido
nx.draw(G, pos=nx.spring_layout(G), with_labels=True, font_weight='bold', font_color="w")

# Mostra o caminho Euleriano do novo Grafo
print("Caminho Euleriano:", list(nx.eulerian_circuit(G)))
```
