```python
def gera_grafo():
    #Definicao de variaveis
    aresta = []
    
    # Abre o arquivo somente leitura
    readfile  = open("grafobipartido.txt", "r")

    #Percorre o arquivo
    for line in readfile:
        aresta.append(tuple(line.rstrip().split(' ')))

    #Cria o Grafo
    G = nx.Graph(aresta)        
        
    #Plota o Grafo
    nx.draw(G, pos=nx.spring_layout(G), with_labels=False, node_color='red', node_size=2, edge_color='orange', width=0.2)
    
    return G

B = gera_grafo()
```
